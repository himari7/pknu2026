```python
from konlpy.tag import Okt
okt = Okt()

positive_words = {
    "행복하다": 2,
    "사랑하다": 2,
    "감사하다": 2,
    "즐겁다": 2,
    "아름답다": 2,
    "감동하다": 1,
    "만족하다": 1,
    "희망하다": 2,
    "존경하다": 2,
    "기쁘다": 2,
    "뿌듯하다": 1,
    "건강하다": 1,
    "성공하다": 2,
    "믿음직하다": 1,
    "따뜻하다": 1,
    "자랑스럽다": 2,
    "용기있다": 1,
    "긍정적이다": 2,
    "밝다": 1,
    "신나다": 2
}
negative_words = {"슬프다": -2, "화나다": -2, "미워하다": -2, "불행하다": -2, "지겹다": -1, "무가치하다": -2, "괴롭다": -2, "싫다": -2, "두렵다": -1, "우울하다": -2,
    "절망하다": -2,
    "불안하다": -1,
    "짜증나다": -1,
    "외롭다": -2,
    "후회하다": -1,
    "비참하다": -2,
    "실망하다": -1,
    "억울하다": -1,
    "초조하다": -1,
    "불편하다": -1
}
text = input("당신의 기분을 한 문장으로 표현해 주세요")

words = okt.morphs(text, stem=True)

i = 0 
point = 0
while i < len(words):
    
    if words[i] in positive_words:
        point += positive_words[words[i]]
    
    if words[i] in negative_words:
        point -= negative_words[words[i]]
    i += 1

print(words)
print(point)
```
